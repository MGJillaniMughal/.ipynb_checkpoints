```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sample=pd.read_csv("android.csv")
sample.head()
```

```python
sample["category"] = sample["category"].astype("category")
```

```python
sample["category"]
```

```python
sample["installs"].sum()
```

```python
sample["5 star ratings"].sum()
```

```python
sa = sample.sort_values(by = ['installs','average rating'], ascending = False).head(200)
```

```python
sa =sample.groupby(by = 'category')['installs'].count().reset_index().sort_values(by = 'installs', ascending = False)
```

```python
sns.lineplot(x= 'installs', y = 'category' ,data = sa)
```
